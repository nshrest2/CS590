```r
# Import peanut_lines.csv and store data as lines
lines <- read.csv('peanut_lines.csv')
```

```r
lines
```

```r
# Show the first 25 rows of lines
head(lines,25)
```

```r
length(grep('Bailey', lines[['Parentage']]))
```

```r
Bailey <- function(x){
    if (grepl('Bailey', x)){
        return('Yes')
    }else{
        return('No')
    }
}
lines['Bailey_Parent'] <- sapply(lines[['Parentage']], Bailey)
```

```r
head(lines,10)
```

```r
# What percentage of lines were developed using 'Bailey' are verified as high-oleic (Fatty Acid Genotype[FAG]='ol ol')?

bailey_percent = round(nrow(lines[lines['Bailey_Parent']=='Yes' & lines['FAG'] == 'ol ol',]) / nrow(lines)*100,1)
noquote(sprintf('%s%%', bailey_percent))
```

```r
# Replace the name 'Bailey' with the tested name 'N03081T' permanently for all instances in the 'Parentage' column.

lines['Parentage']<- gsub('Bailey', 'N03081T', lines[['Parentage']])
```

```r
head(lines,10)
```

```r
# Load the 'peanut_yield.txt' dataset as 'yields'

yields <- read.csv('peanut_yield.txt', sep='\t', header=TRUE)
```

```r
yields
```

```r
# Merge 'lines' and 'yields' together on 'NC_Accession and name it 'peanut_db'
peanut_db <- merge (lines, yields, by.x='NC_Accession', by.y='NC_Accession', all.y=TRUE)
```

```r
peanut_db
```

```r
# Create a dataframe that contains the average yield of Bailey and the average of all other lines with Bailey as the parent

peanut_db <- peanut_db[peanut_db['NC_Accession'] == 'Bailey' | peanut_db['Bailey_Parent']=='Yes',]

```

```r
peanut_db
```

```r
Line <- c('Bailey', 'Other')
Yield <- c(mean(peanut_db[peanut_db['NC_Accession'] =='Bailey',]$Yield, na.rm=TRUE), mean(peanut_db[peanut_db['NC_Accession']!='Bailey',]$Yield, na.rm=TRUE))
df <- data.frame(row.names = Line, Yield)
```
