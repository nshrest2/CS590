```r
# Input the peanut_lines(CSV) and peanut_disease (Tab) datafile

peanut_lines <- read.csv('peanut_lines.csv',header=TRUE)
peanut_disease <- read.table('peanut_disease.txt',sep='\t', header=TRUE)
```

```r
head(peanut_lines)
dim(peanut_lines)
```

```r
head(peanut_disease)
dim(peanut_disease)
```

```r
# Check the column names of peanut_lines and peanut_disease
colnames(peanut_lines)
```

```r
colnames(peanut_disease)
```

```r
# Merge peanut_lines and peanut_disease keeping the entirety of the peanut_disease information intact - Name the data frame disease_data

disease_data <- merge(peanut_lines, peanut_disease, all=TRUE)
```

```r
disease_data
```

```r
# Subset the disease data to list only ratings for leaf spot ('LS') in a dataframe called leaf.spot

head(disease_data)
```

```r
leaf.spot <- subset(disease_data, Rating =='LS')
```

```r
leaf.spot
```

```r
head(leaf.spot)
dim(leaf.spot)
```

```r
# Drop all rows that contain missing values in the column 'Disease'

any(is.na(leaf.spot$Disease))
```

```r
leaf.spot <- leaf.spot[!is.na(leaf.spot$Disease),]
```

```r
any(is.na(leaf.spot$Disease))
```

```r
# What was the average leaf spot rating in 2017?

leaf.spot.2017 <- subset(leaf.spot, Year==2017)
summary(leaf.spot.2017$Disease)
```

```r
# Subset the leaf.spot dataframe to include the following lnes

line_names = c('Bailey','Bailey II','Wynne','Sullivan','Emery','N96076')
line_names
```

```r
leaf.spot2 <- leaf.spot[leaf.spot$NC_Accession %in% line_names,]
```

```r
unique(leaf.spot2$NC_Accession)
```

```r
# Which line is the most leaf spot resistant?

help(aggregate)
```

```r
head(leaf.spot2)
```

```r
aggregate(Disease ~ NC_Accession, data = leaf.spot2, mean)
```
